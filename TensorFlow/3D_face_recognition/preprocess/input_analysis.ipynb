```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```python
rgb_path = os.path.expanduser('../../README/0001_03.jpg')
saved_rgb_path = os.path.expanduser('../../README/rgb_0001_03.jpg')
rgb = mpimg.imread(rgb_path)
plt.imshow(rgb)
mpimg.imsave(saved_rgb_path, rgb)
```

```python
dep_path = os.path.expanduser('../../README/0001_03.npy')
saved_dep_path = os.path.expanduser('../../README/dep_0001_03.jpg')
dep = np.load(dep_path)
plt.matshow(dep)
plt.show()
mpimg.imsave(saved_dep_path, dep)
```
